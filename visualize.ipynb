```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
# fake training and validation loss
train_loss = []
valid_loss = []
for step in range(200):
    if step % 2 == 0:
        train_loss.append((step, np.exp(-1/20 * step) + 0.1 * np.random.rand()))
    if step % 10 == 0:
        valid_loss.append((step, np.exp(-1/40 * step) + 0.1 * np.random.rand() + 0.2))
```

```python
# save npz
print(train_loss[:3])
print(valid_loss[:3])
np.savez(
    'log.npz',
    train_loss = train_loss,
    valid_loss = valid_loss,
)
```

```python
# load npz
log = np.load('log.npz')
train_step, train_loss = list(zip(*log['train_loss']))
valid_step, valid_loss = list(zip(*log['valid_loss']))
```

```python
fig = plt.figure(figsize=(16, 9))
plt.plot(train_step, train_loss)
plt.plot(valid_step, valid_loss)
```
